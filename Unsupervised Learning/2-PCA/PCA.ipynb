# Principal Component Analysis (PCA) on Iris Dataset

## Introduction

Welcome to the PCA project! In this Jupyter notebook, we explore Principal Component Analysis (PCA), a crucial dimensionality reduction technique widely used in data analysis and machine learning. This technique simplifies complex datasets by transforming them into a new set of orthogonal variables, known as principal components, which are invaluable for visualization, noise reduction, feature extraction, and more.

## PCA Algorithm

PCA operates through the following steps:

1. **Standardization**: Standardize the dataset to have zero mean and unit variance for each feature.
2. **Covariance Matrix**: Calculate the covariance matrix of the standardized data.
3. **Eigenvalue and Eigenvector Computation**: Compute the eigenvalues and eigenvectors of the covariance matrix.
4. **Sort and Select Components**: Sort the eigenvalues in descending order and select the top k eigenvectors corresponding to the largest eigenvalues, where k is the desired number of principal components.
5. **Projection**: Project the data onto the selected principal components to obtain a reduced-dimensional representation of the dataset.

## Advantages and Disadvantages

**Advantages**:
- Dimensionality Reduction: Helps in reducing the dataset's dimensionality while preserving most of the variance.
- Noise Reduction: Can remove noise and irrelevant information from the data.
- Visualization: Enables visualization of high-dimensional data in a lower-dimensional space.
- Feature Extraction: Extracts meaningful features or components from the data.

**Disadvantages**:
- Linearity Assumption: Assumes linear relationships between variables and may not perform well for nonlinear data.
- Interpretability: Principal components may not always be straightforward to interpret.
- Loss of Information: Reducing dimensions can result in some loss of information.

## Using PCA for Flower Type Clustering with the Iris Dataset

In this project, we leverage the Iris dataset to cluster flower types based on their sepal and petal measurements. This analysis demonstrates how PCA can facilitate dimensionality reduction and feature extraction, making it an invaluable tool for botanical research and educational purposes.

## Implementation

First, let's reload the Iris dataset and preprocess it for PCA:

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data= iris.data, columns= iris.feature_names)
iris_df['target'] = iris.target

# Standardize the features
scaler = StandardScaler()
iris_standardized = scaler.fit_transform(iris_df.iloc[:, :-1])

# Apply PCA
pca = PCA(n_components=3)
principal_components = pca.fit_transform(iris_standardized)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
colors = ['r', 'g', 'b']

for target, color in zip([0,1,2], colors):
    ax.scatter(principal_components[iris_df['target'] == target, 0],
               principal_components[iris_df['target'] == target, 1],
               principal_components[iris_df['target'] == target, 2],
               c=color, label=f'Class {target}')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.legend()
plt.show()
```
